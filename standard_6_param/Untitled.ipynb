```python
import numpy as np
import matplotlib.pyplot as plt
import os


# --- Main Plotting Code ---

# File names
file1 = 'lensedCls_from_c.dat'
file2 = 'lensedCls_from_c_2.dat'
file3 = 'test_2_lensedCls.dat'
# Load the data from both files using numpy
# It will skip rows that can't be parsed, handy for file headers/footers
data1 = np.loadtxt(file1)
data2 = np.loadtxt(file2)
data3 = np.loadtxt(file3)
# Extract columns for each file
# Column 0 is L, the rest are the power spectra
l1, tt1, ee1, bb1, te1 = data1[:, 0], data1[:, 1], data1[:, 2], data1[:, 3], data1[:, 4]
l2, tt2, ee2, bb2, te2 = data2[:, 0], data2[:, 1], data2[:, 2], data2[:, 3], data2[:, 4]
l3, tt3, ee3, bb3, te3 = data3[:, 0], data3[:, 1], data3[:, 2], data3[:, 3], data3[:, 4]
# Create a figure with 4 subplots, arranged in a 2x2 grid
# figsize is width, height in inches for the whole figure
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Set a main title for the entire figure
fig.suptitle('Comparison of Lensed CMB Power Spectra', fontsize=20)

# --- Plot 1: TT Spectrum ---
ax1 = axes[0, 0]
ax1.plot(l1, tt1, label='File 1: lensedCls_from_c.dat', color='blue')
ax1.plot(l2, tt2, label='File 2: lensedCls_from_c_2.dat', color='red', linestyle='--')
ax1.plot(l3, tt3, label='File 3: test_2_lensedCls.dat', color='green')
ax1.set_title('TT Power Spectrum', fontsize=16)
ax1.set_xlabel('Multipole moment $L$', fontsize=12)
ax1.set_ylabel('$D_L^{TT} \ [\mu K^2]$', fontsize=12)
ax1.legend()
ax1.grid(True, linestyle=':', alpha=0.6)

# --- Plot 2: EE Spectrum ---
ax2 = axes[0, 1]
ax2.plot(l1, ee1, label='File 1', color='blue')
ax2.plot(l2, ee2, label='File 2', color='red', linestyle='--')
ax2.plot(l3, ee3, label='File 3', color='green')
ax2.set_title('EE Power Spectrum', fontsize=16)
ax2.set_xlabel('Multipole moment $L$', fontsize=12)
ax2.set_ylabel('$D_L^{EE} \ [\mu K^2]$', fontsize=12)
ax2.legend()
ax2.grid(True, linestyle=':', alpha=0.6)

# --- Plot 3: BB Spectrum ---
ax3 = axes[1, 0]
ax3.plot(l1, bb1, label='File 1', color='blue')
ax3.plot(l2, bb2, label='File 2', color='red', linestyle='--')
ax3.plot(l3, bb3, label='File 3', color='green')
ax3.set_title('BB Power Spectrum', fontsize=16)
ax3.set_xlabel('Multipole moment $L$', fontsize=12)
ax3.set_ylabel('$D_L^{BB} \ [\mu K^2]$', fontsize=12)
ax3.legend()
ax3.grid(True, linestyle=':', alpha=0.6)

# --- Plot 4: TE Spectrum ---
ax4 = axes[1, 1]
ax4.plot(l1, te1, label='File 1', color='blue')
ax4.plot(l2, te2, label='File 2', color='red', linestyle='--')
ax4.plot(l3, te3, label='File 3', color='green')
ax4.set_title('TE Cross-Correlation Spectrum', fontsize=16)
ax4.set_xlabel('Multipole moment $L$', fontsize=12)
ax4.set_ylabel('$D_L^{TE} \ [\mu K^2]$', fontsize=12)
ax4.legend()
ax4.grid(True, linestyle=':', alpha=0.6)

# Adjust layout to prevent titles and labels from overlapping
plt.tight_layout(rect=[0, 0.03, 1, 0.96]) # Adjust rect to make space for suptitle

# Display the plots
plt.show()

```
